```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.journaldunet.com/management/ville/seine-et-marne/departement-77/villes'
req = requests.get(url)
```

```python
req
```

```python
contenu = req.content
```

```python
soup = bs(contenu, 'html.parser')
```

```python
print(soup)
```

```python
groupeListes = soup.findAll('ul', class_ = 'odSquareList bloc size1of2')
```

```python
len(groupeListes)
```

```python
groupeListes
```

```python
for ul in groupeListes:
    for a in ul.findAll('a'):
        if '/ville-' in a['href']:
            print(a['href'])
    print('fin de la liste')
```
